【例4.13】ベータ・二項分布 $X|Y \sim Bin(n,Y),Y \sim Beta(\alpha,\beta) $

```julia
α=.5; β=.5; n=100; 
surface(0:1:100, 0:.1:1, (x,y) ->  pdf(Binomial(n, y) ,x)*pdf(Beta(α,β),y),color=:bluesreds)
```

【例4.15】ガンマ・ポアソン分布 $X|Y \sim Po(n,Y),Y \sim Ga(\alpha,\beta) $

```julia
α=5; β=3;
surface(0:1:50, 0:1:50, (x,y) ->  pdf(Poisson(y) ,x)*pdf(Gamma(α,β),y),color=:bluesreds)
```

$X$と$Y$は独立に標準正規分布に従う

```julia
surface(-3:0.1:3,-3:0.1:3,(x,y)->pdf(Normal(0,1),x)*pdf(Normal(0,1),y))
```
