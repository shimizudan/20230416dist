【例4.16】$X$と$Y$は独立に標準正規分布に従う。

1. $X=R\cos\Theta$，$Y=R\sin\Theta$によって変数変換を行う。$(X,Y)\to(R,\Theta)$

$f_{R,\Theta}(r,\theta)=\dfrac{re^{-\frac{r^2}2}}{2\pi},f_{R}(r)=re^{-\frac{r^2}2},f_{\Theta}(\theta)=\dfrac{1}{2\pi}$


2. さらに$V=R^2$によって，変数変換を行う。$(R,\Theta)\to(V,\Theta)$

$f_{V,\Theta}(v,\theta)=\dfrac{e^{-\frac{v}2}}{4\pi}$


```julia
p1=surface(-3:0.1:3,-3:0.1:3,(x,y)->pdf(Normal(0,1),x)*pdf(Normal(0,1),y),color=:tab10)
p2=surface(0:0.1:3,0:0.1:π ,(r,θ)->pdf(Normal(0,1),r*cos(θ))*pdf(Normal(0,1),r*sin(θ))*r,color=:tab10)
p3=surface(0:0.1:3,0:0.1:π ,(v,θ)->exp(-v/2)/(4*π),color=:tab10)

plot(p1,p2,p3,layout=(1,3),size=(1200,300))
```

【例4.17】$X$と$Y$は独立で，$X\sim Ga(a,1)$，$Y\sim Ga(b,1)$（$a>0$，$b>0$）

$Z=X+Y$，$W=\dfrac{X}{X+Y}$によって変数変換を行う。$(X,Y)\to(Z,W)$


```julia
a=2; b=3;
p4=surface(0:0.1:10,0:0.1:10,(x,y)->pdf(Gamma(a,1),x)*pdf(Gamma(b,1),y),color=:roma)
p5=surface(0:0.1:10,0:0.1:10 ,(z,w)->pdf(Gamma(a,1),z*w)*pdf(Gamma(b,1),z*(1-w))*z,color=:roma)
plot(p4,p5,layout=(1,2),size=(900,300))
```
